```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
cancer_data = pd.read_csv('/content/data.csv')
```

```python
cancer_data.head()
```

```python
cancer_data=cancer_data.drop(columns='Unnamed: 32')
```

```python
cancer_data.describe()
```

```python
cancer_data.groupby('diagnosis').mean()
```

```python
cancer_data=cancer_data.drop(columns='id')
```

```python
cancer_data.replace({'diagnosis':{'M':1,'B':0}},inplace=True)
```

```python
cancer_data.head()
```

```python
correlation = cancer_data.corr()
```

```python
plt.figure(figsize=(12,12))
sns.heatmap(data = correlation,cmap='Blues',cbar=True,fmt='.1f',annot=True,annot_kws={'size':8})
```

```python
x = cancer_data.drop(columns='diagnosis',axis=1)
y =cancer_data.diagnosis
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
model =LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_predict = model.predict(x_train)
acc = accuracy_score(y_train,y_predict)
print(acc)
```

```python
y_test_predict = model.predict(x_test)
acc = accuracy_score(y_test,y_test_predict)
print(acc)
```
